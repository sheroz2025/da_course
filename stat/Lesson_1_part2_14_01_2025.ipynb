```python
import numpy as np
import random as pd
import matplotlib.pyplot as plt
```

```python
def flip_coins(n):
    results = []
    flips = []
    proportion= []

    for i in range(1, n+1):
        flip= random.randint(0, 1)
        results.append(flip)
        flips.append(i)
        heads_proportion= np.mean(results)
        proportion.append(heads_proportion)
    return flips, proportion
```

```python
flips, proportions = flip_coins(10)
plt.plot(flips, proportions)
```
